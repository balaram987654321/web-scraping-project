```python
import requests
```

```python
webpage=requests.get("https://www.worldometers.info/world-population/population-by-country/")
```

```python
webpage
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(webpage.content)
```

```python
soup
```

```python
details=soup.find_all("div",class_="col-md-12")
```

```python
details
```

```python
info=[]
for i in details:
    for j in i.find_all("td"):
        info.append(j.text.replace("\n",""))
info
```

```python
srno=info[0::12]
srno
```

```python
country=info[1::12]
country
```

```python
population=info[2::12]
population
```

```python
yearlychange=info[3::12]
yearlychange
```

```python
netchange=info[4::12]
netchange
```

```python
density=info[5::12]
density
```

```python
landarea=info[6::12]
landarea
```

```python
migrants=info[7::12]
migrants
```

```python
fertrate=info[8::12]
fertrate
```

```python
medage=info[9::12]
medage
```

```python
urbanpop=info[10::12]
urbanpop
```

```python
worldshare=info[11::12]
worldshare
```

```python
import pandas as pd
```

```python
WORLDPOPULATION=pd.DataFrame({})
WORLDPOPULATION["SR.NO"]=srno
WORLDPOPULATION["COUNTRY NAMES"]=country
WORLDPOPULATION["POPULATION 2020"]=population
WORLDPOPULATION["YEARLY CHANGE"]=yearlychange
WORLDPOPULATION["NET CHANGE"]=netchange
WORLDPOPULATION["DENSITY P/KM^2"]=density
WORLDPOPULATION["LAND AREA KM^2"]=landarea
WORLDPOPULATION["MIGRANTS(NET)"]=migrants
WORLDPOPULATION["FERT.RATE"]=fertrate
WORLDPOPULATION["MED AGE"]=medage
WORLDPOPULATION["URBAN POPULATION"]=urbanpop
WORLDPOPULATION["WORLD SHARE"]=worldshare
WORLDPOPULATION
```

```python

from matplotlib import pyplot as plt
import numpy as np

WORLDPOPULATION[['COUNTRY NAMES','POPULATION 2020']].value_counts()[0:8].plot(kind='pie')
```

```python
WORLDPOPULATION.to_csv
```

```python
import pymysql as bk
```

```python
con1=bk.connect(host="localhost",user="root",database="world population",password="")
print("connection has done successfully")
```

```python
mycursor=con1.cursor()
mycursor.execute("create table if not exists population(SRNO int,COUNTRY varchar(50),POPULATION int,YEARLYCHANGES int,NETCHANGE int,DENSITY int,LANDAREA int,MIGRANTS int,FERTRATE int,MEDAGE int,URBANPOPULATION int,WORLDSHARE int)")
print("table created successfully")
```

```python
for i in range(len(srno)):
    mycursor.execute("""insert into population values("%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s")"""%(srno[i],country[i],population[i],yearlychange[i],netchange[i],density[i],landarea[i],migrants[i],fertrate[i],medage[i],urbanpop[i],worldshare[i]))
    con1.commit()
```
